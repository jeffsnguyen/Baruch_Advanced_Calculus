```python
import numpy as np
import scipy.stats as spstats
#######################

#######################
# Return f(sigma_k) value
def fn(S_0, K, r, q, T, sigma, V_m, cpFlag):
    d1 = (np.log(S_0/K) + (r - q + 0.5*(sigma**2)) * T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    N_d1 = spstats.norm.cdf(d1,0,1)
    N_d2 = spstats.norm.cdf(d2,0,1)

    # If call-put flag = "c" then return C - V_m
    if cpFlag == "c":
        return S_0*N_d1 - K*np.exp(-r*T)*N_d2 - V_m
    else: # else return P - V_m
        return K*np.exp(-r*T) - S_0 + S_0*N_d1 - K*np.exp(-r*T)*N_d2 - V_m

#######################
# Return vega(sigma_k) value
def vegaFunc(S_0, K, r, q, T, sigma):
    d1 = (np.log(S_0/K) + (r - q + 0.5*(sigma**2)) * T) / (sigma*np.sqrt(T))
    N_d1 = spstats.norm.cdf(d1,0,1)
    
    return S_0*np.exp(-q*T)*N_d1*np.sqrt(T)

# Newton estimation
def Newton(S_0, K, r, q, T, sigma, V_m, flag):
    fnVal = fn(S_0, K, r, q, T, sigma, V_m, flag)
    vegaVal = vegaFunc(S_0, K, r, q, T, sigma)
    
    return sigma - (fnVal/vegaVal)

# Estimating with tolerance
def tolCheck(S_0, K, r, q, T, sigma, V_m, flag, tol):
    k=0
    prev = Newton(S_0, K, r, q, T, sigma, V_m, flag)
    
    while True:
        k+=1
        current = Newton(S_0, K, r, q, T, prev, V_m, flag)
        if abs(current-prev) > tol:
            print(f'sigma_{k} = {current}')
            prev = current
        else:
            print(f'Converged at sigma_{k} = {current} and tolerance = {abs(current-prev)} <= {tol}')
            break
    
    return current
    
#######################
# Given values
S_0 = 50
K = 45
r = 1.5/100
q = .5/100
T = 9/12
V_m = 8.5

# Given method-specific parameters
sigma_0 = .2
flag="c"

# Set tolerance
tol = 10**(-8)

sigma = tolCheck(S_0, K, r, q, T, sigma_0, V_m, flag, tol)
```

```python
import numpy as np
import scipy.stats as spstats

# Return Put value at time 0 using Black Scholes
def P0_BS(S_0, K, r, q, T, sigma):
    d1 = (np.log(S_0/K) + (r - q + 0.5*(sigma**2)) * T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    N_minusd1 = spstats.norm.cdf(-d1,0,1)
    N_minusd2 = spstats.norm.cdf(-d2,0,1)
    return K*np.exp(-r*T)*N_minusd2 - S_0*np.exp(-q*T)*N_minusd1

#######################
def func(S_0, K, r, q, T, sigma):
    d1 = (np.log(S_0/K) + (r - q + 0.5*(sigma**2)) * T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    N_minusd1 = spstats.norm.cdf(-d1,0,1)
    N_minusd2 = spstats.norm.cdf(-d2,0,1)
    
    return K*np.exp(-r*T)*N_minusd2 - S_0*np.exp(-q*T)*N_minusd1 - K + S_0

#######################
def funcPrime(S_0, K, r, q, T, sigma):
    d1 = (np.log(S_0/K) + (r - q + 0.5*(sigma**2)) * T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    N_minusd1 = spstats.norm.cdf(-d1,0,1)
    N_minusd2 = spstats.norm.cdf(-d2,0,1)
    
    return np.exp(-r*T)*N_minusd2 + (1/(sigma*np.sqrt(2*np.pi*T)))*np.exp(-r*T-.5*d2**2) - (1/(K*np.sqrt(2*np.pi)))*S_0*np.exp(-q*T-.5*d1**2) - 1

# Estimating with tolerance
def tolCheck(S_0, K_0, r, q, T, sigma, tol):
    k=0
    prev = K_0
    
    while True:
        k+=1
        current = prev - func(S_0, prev, r, q, T, sigma)/funcPrime(S_0, prev, r, q, T, sigma)
        if abs(current-prev) > tol:
            print(f'K_{k} = {current}')
            prev = current
        else:
            print(f'Converged at K_{k} = {current} and tolerance = {abs(current-prev)} <= {tol}')
            break
    
    return current
# Given values
S_0 = 50
K_0 = 50
r = .03
q = .01
T = 6/12
sigma = .3
tol = 10**(-6)

tolCheck(S_0, K_0, r, q, T, sigma, tol)
```
